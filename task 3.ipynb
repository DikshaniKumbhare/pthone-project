# Cumulative Distribution Function (CDF)

```python
yof = np.array(df["math score"])
counts, bin_edges = np.histogram(yof,bins = 10, density = True)
pdf = counts/sum(counts)
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label = "pdf")
plt.plot(bin_edges[1:],cdf, label = "cdf")
plt.legend()
plt.xlabel("math score")
plt.ylabel("pdf / cdf")
plt.title("PDF and CDF of 'math score' ")

plt.show()
```

# Observations
1. This plot shows the PDF and CDF on "maths score" feature.
2. PDF shows percentage of patients having x is maths score where x indicates the bins or range of numbers
3. As one can notice from PDF that, most of the score are with 10 to 50 . 
4. This can be observed in case of CDF also as it is becoming constant for score in range 20 to 40.
5. PDF - Almost 80% students have got 65 marks in maths , 10% students have above 65 marks in maths
6. CDF- almost 10% students got 40 marks in maths.

5. Box says 50 % points lies in that region because it shows between 25th to 75th percentile range

```python
Q1 =  df['math score'].quantile(0.25)
Q3 = df['math score'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```
